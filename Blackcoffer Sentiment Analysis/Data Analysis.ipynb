```python
def complex_word_count(word):
    word = word.lower()
    count = 0
    vowels = 'aeiou'
    if word[0] in vowels:
        count += 1
    for index in range(1,len(word)):
        if word[index] in vowels and word[index-1] not in vowels:
            count += 1
    if count == 0:
        count += 1
    return count
```

```python
path1 = r'C:\Users\chhot\Dropbox\BlackCoffer\text files'
path2 = r'C:\Users\chhot\Dropbox\BlackCoffer\text file cleaned'

tfiles = os.listdir(path1)
wrdcnt = []
cwrdcnt = []
twac = []
for i in dataframe['URL_ID'].values:
    with open (r'{}\{}.{}'.format(path1,i,'txt'), encoding='utf-8') as infile, open(r'{}\{}.{}'.format(path2,i,'txt'),'w',encoding='utf-8') as outfile:
        text = word_tokenize(infile.read())
        wrdcnt.append(len(text))
        cwcnt = 0
        for wrd in text:
            if complex_word_count(wrd)>2:
                cwcnt += 1
        cwrdcnt.append(cwcnt)
        print(cwrdcnt)
        filetext = [txt for txt in text if not txt in stopwords]
        outfile.write(' '.join(filetext))
        twac.append(len(filetext))

dataframe['wrdcnt'] = wrdcnt
dataframe['COMPLEX WORD COUNT'] = cwrdcnt
dataframe['twac'] = twac
```

```python
with open(r'{}\{}'.format(path4,'positive-words.txt'), encoding='windows-1252') as file:
    PositiveWords = word_tokenize(file.read())

with open(r'{}\{}'.format(path4,'negative-words.txt'), encoding='windows-1252') as file:
    NegativeWords = word_tokenize(file.read())
```

```python
positive_score = []
for i in dataframe['URL_ID'].values:
    pscore = 0
    with open(r'{}\{}.{}'.format(path2,i,'txt'),encoding='utf-8')as f:
        for txt in word_tokenize(f.read()):
            for txt1 in PositiveWords:
                if txt == txt1: pscore+=1
    positive_score.append(pscore)
dataframe["POSITIVE SCORE"] = positive_score
```

```python
negative_score = []
for i in dataframe['URL_ID'].values:
    nscore = 0
    with open(r'{}\{}.{}'.format(path2,i,'txt'), encoding='utf-8') as f:
        for txt in word_tokenize(f.read()):
            for txt1 in NegativeWords:
                if txt == txt1: nscore-=1
    negative_score.append(-nscore)
dataframe['NEGATIVE SCORE'] = negative_score
```

```python
dataframe['POLARITY SCORE']= (dataframe['POSITIVE SCORE']-dataframe['NEGATIVE SCORE'])/((dataframe['POSITIVE SCORE']+dataframe['NEGATIVE SCORE'])+ 0.000001)
```

```python
dataframe['SUBJECTIVITY SCORE'] = (dataframe['POSITIVE SCORE']+dataframe['NEGATIVE SCORE'])/((dataframe['twac'])+ 0.000001)
```

```python
def syllable_count(word):
    word = word.lower()
    count = 0
    vowels = "aeiou"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index-1] not in vowels:
            count += 1
    if word.endswith(('es','ed')):
        count -= 1
    if count == 0:
        count += 1
    return count
```

```python
scl = []
for i in dataframe['URL_ID'].values:
    with open(r'{}\{}.{}'.format(path1,i,'txt'),encoding='utf-8') as f:
        words = word_tokenize(f.read())
        sc = 0
        for word in words:
            sc += syllable_count(word)
    scl.append(sc)
    print(scl)
```
