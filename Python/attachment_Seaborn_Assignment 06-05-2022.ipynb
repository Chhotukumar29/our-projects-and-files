```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
cars=pd.read_csv('cars.csv')
```

```python
cars.head()
```

### reproduce the following plots with appropriate code

```python
sns.boxplot(cars.Actual_MPG)
```

```python
sns.violinplot(cars.Actual_MPG)
```

### univariate distributions

```python
sns.distplot(cars.Actual_MPG,bins=20,kde=False)
# help(sns.distplot)
```

```python
sns.distplot(cars.Actual_MPG,bins=20,hist=False)
```

```python
sns.distplot(cars.Actual_MPG,bins=20,kde=False, rug=True)
# help(sns.distplot)
```

```python
sns.distplot(cars.Actual_MPG,bins=20, rug=True)
```

###  bivariate distributions

```python
sns.jointplot(cars.Actual_MPG,cars.Horsepower)
```

```python
sns.jointplot(cars.Actual_MPG,cars.Horsepower,kind='kde')
```

```python
sns.pairplot(cars)
```

```python
sns.PairGrid(cars).map_diag(sns.kdeplot).map_offdiag(sns.kdeplot, cmap="Blues_d")
# help(sns.pairplot)
```
